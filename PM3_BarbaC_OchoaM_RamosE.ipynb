# Ecuaciones de Lotka-Volterra: Modelo Presa- Depredador

Utilizar la ecuación diferencial de lotka-volterra para simular el resultado de la cadena alimenticia  

## Objetivos

### Objetivo General 

Realizar un modelo presa-depredador, en este caso sería de leones-antílopes, donde podamos observar como conforme decrece el número de presas, la población de depredadores decrece en algun momento como resultado de un menor suministro de alimento, y con un decrecimiento en el número de depredadores hay un incremento en el número de presas; y así observar como es la relación del incremento y decrecimineto de cada uno 

### Objetivos específicos

- Resolver la ecuación diferencial
- Graficar la población de presa-depredador
- Analizar como queda la cadena alimenticia

## Problema a resolver

Modelo presa-depredador

$$\frac{dx}{dt}= -0.16x + 0.08xy$$ (presa) (antilopes)

$$\frac{dy}{dt}= 4.5y - 0.9xy$$ (depredador) (leones)

Debido a que se está tratando con poblaciones, se tiene x(t)>0 y(t)>0. Las condiciones iniciales que se utilizan fueron x(0)=5, y(0)=5

## Resolución

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from scipy.integrate import odeint
```

```python
def lotkaVolterra(xt, t):
    x = xt[0]
    y = xt[1]
    dx = (-.16*x)+(.08*x*y)
    dy = (4.5*y)-(.9*x*y)
    return np.array([dx, dy])
```

```python
x0 = np.array([5, 5])
t = np.linspace(0, 25, 100)
print(x0)
```

```python
xx = odeint(lotkaVolterra, x0, t)
x1 = xx[:, 0]
x2 = xx[:, 1]
```

```python
plt.figure(figsize=(8,8))
plt.plot(t, x1, label='Población de antilopes $x_1(t)$')
plt.plot(t, x2, label='Población de leones $x_2(t)$')
plt.xlabel('Tiempo $t$')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.legend(loc='best')
plt.grid()
plt.show()
```

Vemos que se trata de una solución periódica en la que, un aumento en la población de antilopes va seguido del aumento del número de leones. Un gran número de depredadores merma la población de presas y a los pobres leones les toca pasar hambre una temporada

```python
plt.figure(figsize=(6,6))
plt.plot(x1, x2, label='Trayectoria de las poblaciones')
plt.plot(x1[0], x2[0], '*', ms=10, label = 'Poblaciones iniciales')
plt.plot(x1[-1], x2[-1], '*', ms=10, label = 'Poblaciones finales')
plt.xlabel('Población de antilopes $x$')
plt.ylabel('Población de leones $y$')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.legend(loc = 'best')
plt.grid()
plt.show()
```

Esta es otra forma de visualizar los datos, aqui se ve el número de presas en función del número de depredadores,es decir, que si recorremos esa curva nos representa la evolución a lo largo del tiempo que se observo en la anterior 

## Conclusiones 

Podemos concluir que conforme el número de leones aumenta, la población de antílopes decrece, y cuando sucede esto al pasar el tiempo los leones comienzan a pasar hambre lo cual provoca que empiecen a morir, y aumentan el número de antílopes, volviéndose un ciclo, sin embargo en nuestro modelo, nos dice que el final de esta cadena termina con más antílopes y menos leones

Este modelo es bastante interesante ya que no solo se puede llevar a ejemplos de cadena alimenticia, dentro de nuestra carrera podríamos utilizarlo para ver como se comportan los vendedores y las compradores en el mercado, utilizando esas variables como si fueran (presa-depredador) y así observar como fluctúan entre ellas, su relación, para poder observar como va el comercio y cuales son sus etapas.
