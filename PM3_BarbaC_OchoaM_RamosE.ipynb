# Ecuaciones de Lotka-Volterra: Modelo Presa- Depredador

Nuestro proyecto consiste en realizar un modelo presa-depredador, en este caso sería de leones-antilopes, donde podamos observar como conforme decrece el numero de presas, la población de depredadores decrece en algun momento como resultado de un menos suministro de alimento, y junto con un decrecimiento en el numero de depredadores hay un incremento en el numero de presas; esto a su vez va a dar lugar a un mayor numero de depredadores, que en una última instancia origina otro decrecimiento en el numero de presas.

## Objetivo General 

Relizar el modelo presa-depredador mostrando como queda en un final la cadena alimenticia

### Objetivos específicos

- Resolver la ecuación diferencial
- Graficar la población de presa-depredador
- Analizar como queda la cadena alimenticia

## Problema a resolver

Modelo presa-depredador

$$dx/dt= -0.16x + 0.08xy$$ (presa) (antilopes)

$$\frac{dy}{dt}= 4.5y - 0.9xy$$ (depredador) (leones)

Debido a que se está tratando con poblaciones, se tiene x(t)>0 y(t)>0. Las condiciones iniciales que se utilizan fueron x(0)=5, y(0)=5

## Resolución

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from scipy.integrate import odeint
```

```python
def lotkaVolterra(xt, t):
    x = xt[0]
    y = xt[1]
    dx = (-.16*x)+(.08*x*y)
    dy = (4.5*y)-(.9*x*y)
    return np.array([dx, dy])
```

```python
x0 = np.array([5, 5])
t = np.linspace(0, 25, 100)
print(x0)
```

```python
xx = odeint(lotkaVolterra, x0, t)
x1 = xx[:, 0]
x2 = xx[:, 1]
```

```python
plt.figure(figsize=(8,8))
plt.plot(t, x1, label='Población de antilopes $x_1(t)$')
plt.plot(t, x2, label='Población de leones $x_2(t)$')
plt.xlabel('Tiempo $t$')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.plot(x1, x2, label='Trayectoria de las poblaciones')
plt.plot(x1[0], x2[0], '*', ms=10, label = 'Poblaciones iniciales')
plt.plot(x1[-1], x2[-1], '*', ms=10, label = 'Poblaciones finales')
plt.xlabel('Población de antilopes $x$')
plt.ylabel('Población de leones $y$')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.legend(loc = 'best')
plt.grid()
plt.show()
```
